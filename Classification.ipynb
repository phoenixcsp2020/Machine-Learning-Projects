# Classification

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

## 1. Data loading

```python
# Loading dataset from sklearn library
data = load_breast_cancer()
X = data.data
y = data.target
```

```python
# Convert to DataFrame 
df1 = pd.DataFrame(X, columns=data.feature_names)
df1['target'] = y
df1
```

```python
# To find out of missing values
print("Missing values in the dataset:\n",df1.isnull().sum())
```

```python
# Since there are no missing values, we proceed with feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#Split data into training and testing
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.2,random_state=0)
```

```python
# Explanation:
#  We scaled the features because the features have different ranges, 
#  and some models (e.g., SVM and k-NN, logistic regression) are sensitive to the scale of the data.
```

## 2. Classification Algorithm Implementation

```python
# 1. Logistic Regression
lr=LogisticRegression(random_state=42)
lr.fit(X_train, y_train)
y_pred_lr=lr.predict(X_test)
#lr

#Formula for accuracy:
#    Accuracy = Number of Correct Predictions/Total Number of Predictions

acc_lr=accuracy_score(y_test, y_pred_lr)
print("Accuracy: ",round(acc_lr,4))
```

```python
# 2. Decision Tree Classifier
dt= DecisionTreeClassifier(random_state=0)
dt.fit(X_train,y_train)
y_pred_dt=dt.predict(X_test)
#dt
acc_dt=accuracy_score(y_test,y_pred_dt)
print("Accurancy: ", round(acc_dt,4))
```

```python
# 3. Random Forest Classifier 
rf=RandomForestClassifier(random_state=0)
rf.fit(X_train,y_train)
y_pred_rf=rf.predict(X_test)
#rf
acc_rf=accuracy_score(y_test,y_pred_rf)
#print("Accuracy: ",round(accuracy_score(y_test,y_pred_rf),4))
print("Accuracy: ",round(acc_rf,4))
```

```python
# 4. Support Vector Machine 
svm=SVC(random_state=0)
svm.fit(X_train,y_train)
y_pred_svm=svm.predict(X_test)
#svm
acc_svm=accuracy_score(y_test,y_pred_svm)
print("Accuracy: ",round(acc_svm,4))
```

```python
# 5. k-Nearest Neighbourab 
knn=KNeighborsClassifier()
knn.fit(X_train,y_train)
y_pred_knn=knn.predict(X_test)
#knn
acc_knn=accuracy_score(y_test,y_pred_knn)
print("Accuracy: ",round(acc_knn,4))
```

## 3. Model Comparison

```python
# Evaluate the performance of each model
print("Accuracy of Logistic Regression: ", acc_lr)
print("Accuracy of Decision Tree: ", acc_dt)
print("Accuracy of Random Forest: ", acc_rf)
print("Accuracy of Support Vector Machine: ", acc_svm)
print("Accuracy of k-Nearest Neighbors: ", acc_knn)
print('-'*50)
```

## Identify the best and worst performing models based on accuracy

```python
print("Best Performing Model is SVM with accuracy of",round(acc_svm,4))
print("Worst Performing Model is Decision Tree with accuracy of",round(acc_dt,4))
```
